```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"C:\Users\zbook 17 g3\Desktop\CV PROJECTS\24.09.2023_19.54.41_REC.png")
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
canny_edged_img=cv2.Canny(img_gray,40,255)
plt.imshow(cv2.cvtColor(canny_edged_img,cv2.COLOR_BGR2RGB))
```

```python
lines=cv2.HoughLinesP(canny_edged_img,1,np.pi/180,30,maxLineGap=250)
```

```python
if lines is not None:
    for line in lines:
        xi,yi,xe,ye=line[0]
        cv2.line(img,(xi,yi),(xe,ye),(255,0,0),3)
        
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
